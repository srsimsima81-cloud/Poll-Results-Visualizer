```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/poll_data.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Choice"].value_counts()
```

```python
plt.figure(figsize=(5,4))
df["Choice"].value_counts().plot(kind="bar")
plt.title("Poll Distribution")
plt.show()
```

```python
pd.crosstab(df["Region"], df["Choice"])
```

```python
pd.crosstab(df["Region"], df["Choice"]).plot(kind="bar", stacked=True)
plt.title("Region-wise Poll Analysis")
plt.show()
```

```python
pd.crosstab(df["Age_Group"], df["Choice"])
```

```python
print("Most popular choice:", df["Choice"].value_counts().idxmax())
print("Least popular choice:", df["Choice"].value_counts().idxmin())
```
